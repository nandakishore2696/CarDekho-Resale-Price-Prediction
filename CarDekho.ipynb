## Handling Data Types

```python
# changing df['KM_TRAVELLED'] from object to numeric
df['KM_TRAVELLED'] = pd.to_numeric(df['KM_TRAVELLED'].str.replace(',',''))
```

```python
# Converting price to numeric
import re 

def extract_price(price):
    num = re.search(r'[\d.]+',price).group()
    return float(num)

df['PRICE'] = df['PRICE'].apply(extract_price)
```

```python
df['REG_YEAR'] = pd.to_numeric(df['REG_YEAR'], errors='coerce')
```

```python
df['SEATS'] = df['SEATS'].astype('int')
```

```python
def extract_engine(text):
    num = re.sub(r' CC','',text)
    try:
        eng = int(num)
    except:
        eng = 0
    return eng

df['ENGINE'] = df['ENGINE'].apply(extract_engine)
```

```python
def extract_mileage(text):
    num = re.search(r'[\d.]+',text).group()
    return float(num)

df['mileage'] = df['mileage'].apply(extract_price)
```

```python
df['COLOR'] = df['COLOR'].str.lower()
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder, OrdinalEncoder
```

```python
onehot_columns =['CITY' , 'FUEL' , 'TYPE' , 'TRANSMISSION' , 'INSURANCE']
ordinal_col = ['OEM', 'MODEL', 'COLOR']

ct = ColumnTransformer(transformers=[('onehot', OneHotEncoder(drop='first', sparse=False), onehot_columns ), ('ordinal', OrdinalEncoder(), ordinal_col)], remainder='passthrough')
ct.set_output(transform='pandas')
df1 = ct.fit_transform(df)
df1
```

```python
df1.columns
```

## Handling Outliers and Skewness

```python
import matplotlib.pyplot as plt 
import seaborn as sns 

columns = ['ordinal__OEM', 'ordinal__MODEL', 'ordinal__COLOR','remainder__KM_TRAVELLED', 'remainder__OWNER_NO', 'remainder__REG_YEAR', 'remainder__SEATS', 'remainder__ENGINE','remainder__mileage']
for col in columns:
    plt.figure(figsize=(10,5))
    plt.subplot(1,2,1)
    sns.kdeplot(df1[col], shade=True)
    plt.title("KDE Plot")

    plt.subplot(1,2,2)
    sns.boxplot(df1[col])
    plt.title("BOX Plot")
    plt.show()
```

```python
# skewed - 
columns_skewed = ['remainder__KM_TRAVELLED', 'remainder__OWNER_NO', 'remainder__SEATS', 'remainder__ENGINE','remainder__mileage']

for i in columns_skewed:
    df1[i] = np.log(df1[i])
```

```python
for col in columns_skewed:
    plt.figure(figsize=(10,5))
    plt.subplot(1,2,1)
    sns.kdeplot(df1[col], shade=True)
    plt.title("KDE Plot")

    plt.subplot(1,2,2)
    sns.boxplot(df1[col])
    plt.title("BOX Plot")
    plt.show()
```
